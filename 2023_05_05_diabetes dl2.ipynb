```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.utils import all_estimators
import warnings
warnings.filterwarnings('ignore')
```

```python
diabetes = pd.read_csv("C:/Users/진현킴/Downloads/diabetes.csv")
```

```python
diabetes
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for i in diabetes.columns :
    diabetes[i] = le.fit_transform(diabetes[i])
```

```python
 diabetes
```

```python
y = diabetes['Outcome']

y.head()
```

```python
X = diabetes.iloc[:,0:8].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=42)
```

```python
import numpy as np

mean = np.mean(X_train, axis = 0)

std = np.std(X_train, axis = 0)

X_train = (X_train - mean) / std
X_test = (X_test - mean) / std
```

```python
print(np.any(np.isnan(X_train)))
print(np.any(np.isnan(X_test)))
print(np.any(np.isinf(X_train)))
print(np.any(np.isinf(X_test)))
```

```python
X_train = pd.DataFrame(X_train).fillna(0).values
X_test = pd.DataFrame(X_test).fillna(0).values
```

```python
print(np.any(np.isnan(X_train)))
print(np.any(np.isnan(X_test)))
print(np.any(np.isinf(X_train)))
print(np.any(np.isinf(X_test)))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(64, activation = 'relu', input_shape = (8,)))
model.add(Dropout(0.5))
model.add(Dense(32, activation = 'tanh'))
model.add(Dropout(0.5))
model.add(Dense(2, activation = 'softmax')) 

model.compile(optimizer = 'rmsprop', loss='mse', metrics = ['mse'])

model.summary()
```

```python
N_EPOCHS = 100
history = model.fit(X_train, y_train, epochs = N_EPOCHS, validation_data=(X_test, y_test))
```
